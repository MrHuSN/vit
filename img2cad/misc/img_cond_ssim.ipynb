```python
import os
import random
import numpy as np
import matplotlib.pyplot as plt

from skimage import io
from skimage.metrics import structural_similarity as ssim
```

```python
img_dir = '../results/week_0101/model_0_10000/1'
img_files = os.listdir(img_dir)
```

```python
ref_img = io.imread(os.path.join(img_dir, '0_cond_img.png'))[:,:,0]
```

```python
def show_img(img):
    plt.imshow(img, cmap='gray')
    plt.axis('off')
    fig = plt.gcf()
    fig.set_size_inches(2, 2)
```

Reference image

```python
show_img(ref_img)
```

Run SSIM between reference and sampled images

```python
comp_imgs = [io.imread(os.path.join(img_dir, img_f))[:,:,0]
    for img_f in img_files if 'cond' not in img_f]
scores = [ssim(ref_img, comp_img) for comp_img in comp_imgs]
```

```python
# Sort by score and display images
idxs = np.argsort(scores)
for sim_idx, idx in enumerate(list(reversed(idxs))):
    print(sim_idx+1, ": ", scores[idx])
    show_img(comp_imgs[idx])
    plt.show()
plt.close()
```
